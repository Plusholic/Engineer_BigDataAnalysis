```python
import pandas as pd

X_test = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/stellar_X_test.csv")
X_train = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/stellar_X_train.csv")
y_train = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/stellar_y_train.csv")
```

```python
train = pd.concat([X_train, y_train], axis=1)
```

```python
train.info()
```

```python
train.describe() # 이상치 제거 해줄지 말지
```

```python
categoric_ = ['run_ID', 'rerun_ID', 'cam_col', 'field_ID', 'plate']
numeric_ = ['alpha','delta','u','g','r','i','z','redshift']
for col in numeric_: # 수치형 변수 이상치 제거
  train = train[(train[col] > train[col].mean() - 2 * train[col].std()) & (train[col] < train[col].mean() + 2 * train[col].std())]
```

```python
train
```

```python
 for col in categoric_:
   print(train[col].value_counts())

# field_ID는 카테고리가 너무 많으니까 제거

del train['field_ID']
del X_test['field_ID']
```

```python
X_train = train.iloc[:,:-1]
y_train = train.iloc[:,-1]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
for col in numeric_:
  X_train[[col]] = scaler.fit_transform(X_train[[col]]) # 두개씩 해주는거 중요
  X_test[[col]] = scaler.fit_transform(X_test[[col]])
```

```python
from sklearn.model_selection import train_test_split # model_seleciton 외우기

X_tr, X_val,y_tr,y_val = train_test_split(X_train, y_train, stratify=y_train)
```

```python
from xgboost import XGBClassifier #대문자
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score

xgb = XGBClassifier()
xgb.fit(X_tr,y_tr)

pred = xgb.predict_proba(X_val)
score = roc_auc_score(y_val, pred[:,1])
print(score)
```

```python
pd.DataFrame({'name' : X_tr.columns, 'imp' : xgb.feature_importances_}).sort_values(by='imp')
```

```python
pred = xgb.predict_proba(X_test)[:,1]
pd.DataFrame({'index' : X_test.index, 'target' : pred})
```

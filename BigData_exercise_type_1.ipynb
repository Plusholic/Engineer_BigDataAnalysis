```python
import pandas as pd
import numpy as np

# 1유형 1번문제
data = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/iris.csv")

# 1유형 1-1
a = data['petal_width'][data['species'] == 'setosa'].quantile(0.75)
b = data['petal_width'][data['species'] == 'virginica'].quantile(0.75)
c = data['petal_width'][data['species'] == 'versicolor'].quantile(0.75)
print(max([a,b,c]))
print(min([a,b,c]))

# 아 이거 조심해야할듯 정리필수. print 해 본 다음에 직접 뺄것.
print('1번 답 : ', max([a,b,c]) - min([a,b,c])) 

# 1유형 1-2

d = data['sepal_length'].value_counts().keys().tolist()
d.sort() # list 오름차순 정렬
print('2번 답 : ', d[9]) # 10번째 오는 숫자

# 1유형 1-3
reverse=False
idx = data['sepal_width'].sort_values(ascending=False).index[:100]
train_data = data.loc[idx]
print('3번 답 : ', int(np.std(train_data['sepal_width'])*50)) #소수점 이하 값은 버림

# 1유형 1-4
a = data['petal_length'][data['species'] == 'setosa'].sort_values(ascending=False)
b = data['petal_length'][data['species'] == 'virginica'].sort_values(ascending=False)
c = data['petal_length'][data['species'] == 'versicolor'].sort_values(ascending=False)
print('4번 답 : ', round(a[:10].mean() + b[:10].mean() + c[:10].mean()))

# 1유형 1-5
data['sepal_length'].describe()
lb = data['sepal_length'].mean() - 2 * data['sepal_length'].std()
ub = data['sepal_length'].mean() + 2 * data['sepal_length'].std()
print('5번 답 : ', len(data[(data['sepal_length'] >= ub) | (data['sepal_length'] <= lb)]))

print(' ')
# 1유형 2번문제

from sklearn.impute import SimpleImputer

data2 = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/wine.csv")

data_1 = data2.copy()
# alcalinity_of_ash 결측치 중앙값 대체
imputer = SimpleImputer(strategy='median')
data_1[['alcalinity_of_ash']] = imputer.fit_transform(data_1[['alcalinity_of_ash']])
# total_phenols  결측치 제거
data_1.dropna(subset=['total_phenols'], inplace=True) # subset 옵션을 지정해줘서 특정 행의 결측치 기준으로 제거해줄수 있음
print('1번 답 : ', int(data_1[['alcalinity_of_ash']].mean()))

# 2-2번
data_2 = data2.copy()
data_2['class'][(data_2['alcohol'] >= data_2['alcohol'].mean()) & (data_2['color_intensity'] >= data_2['color_intensity'].mean())].value_counts()
print('2번 답 : ',0)

# 2-3번
data_3 = data2.copy()
a = data_3['ash'][data_3['class'] == 1].mean()
data_3.dropna(inplace=True)
b = data_3['ash'].mean()
print('3번 답 : ', round(abs(a-b),3))

# 2-4번 #?? 왜이래
data_4 = data2.copy()
# alcohol 변수의 상위 20번째 값으로 상위 20개 값 변환
data_4.sort_values(by='alcohol',ascending=True, inplace=True)
data_4.loc[:19,'alcohol'] = data_4.loc[19,'alcohol']
# proanthocyanins가 2 이하인 데이터를 추출하여 alcohol의 평균값 계산
c = data_4['alcohol'][data_4['proanthocyanins'] <= 2].mean()
print(data_4.head(25))
print('4번 답 : ', round(c,2))

# 책 답
data_4 = data2.copy()
data_sorted = data_4.sort_values(by='alcohol')
data_sorted.loc[0:19,'alcohol'] = data_sorted.loc[19,'alcohol']
print(data_sorted.head(22))

# 2-5번

data_5 = data2.copy()

# class 1에서 proline의 3사분위수 - 1사분위수
d = data_5['proline'][data_5['class'] == 1].quantile(1/4) - data_5['proline'][data_5['class'] == 1].quantile(3/4)
# class 2에서 proline의 3사분위수 - 1사분위수
e = data_5['proline'][data_5['class'] == 2].quantile(1/4) - data_5['proline'][data_5['class'] == 2].quantile(3/4)
print('5번 답 : ', abs(d-e))

```

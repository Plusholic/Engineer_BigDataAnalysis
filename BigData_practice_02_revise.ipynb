```python
import pandas as pd
X_train = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/penguin_X_train.csv")
y_train = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/penguin_y_train.csv")
X_test = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/penguin_X_test.csv")

train = pd.concat([X_train, y_train], axis=1)
train.dropna(inplace=True)
train.reset_index(drop=True, inplace=True)
train = pd.get_dummies(train)

# train.describe()
y_train = train['body_mass_g']
del train['body_mass_g']
X_train = train


from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state = 42)

from sklearn.preprocessing import MinMaxScaler

numeric_ = ['bill_length_mm',	'bill_depth_mm',	'flipper_length_mm']
scaler = MinMaxScaler()

X_tr[numeric_] = scaler.fit_transform(X_tr[numeric_])
X_val[numeric_] = scaler.transform(X_val[numeric_])

from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import roc_auc_score, mean_squared_error

model_xgb = XGBRegressor(random_state = 42, n_estimators=100, max_depth = 2)
model_lgbm = LGBMRegressor(random_state = 42, n_estimators=100, max_depth = 2)
model_rf = RandomForestRegressor(random_state = 42, n_estimators=100, max_depth = 2)
model_lr = LinearRegression()

model_xgb.fit(X_tr,y_tr)
model_lgbm.fit(X_tr,y_tr)
model_rf.fit(X_tr,y_tr)
model_lr.fit(X_tr,y_tr)

pred_xgb = model_xgb.predict(X_val)
pred_lgbm = model_lgbm.predict(X_val)
pred_rf = model_rf.predict(X_val)
pred_lr = model_lr.predict(X_val)

print('*'*30)
mse_xgb = mean_squared_error(y_val, pred_xgb, squared=True)
mse_lgbm = mean_squared_error(y_val, pred_lgbm, squared=True)
mse_rf = mean_squared_error(y_val, pred_rf, squared=True)
mse_lr = mean_squared_error(y_val, pred_rf, squared=True)

rmse_xgb = mean_squared_error(y_val, pred_xgb, squared=False)
rmse_lgbm = mean_squared_error(y_val, pred_lgbm, squared=False)
rmse_rf = mean_squared_error(y_val, pred_rf, squared=False)
rmse_lr = mean_squared_error(y_val, pred_rf, squared=False)
print('RMSE : {0:4f} {1:4f} {2:4f} {3:4f}'.format(mse_xgb, mse_lgbm, mse_rf, mse_lr))
print('MSE : {0:4f} {1:4f} {2:4f} {3:4f}'.format(rmse_xgb, rmse_lgbm, rmse_rf, rmse_lr))
# import sklearn
# help(sklearn.metrics)
```

```python
import pandas as pd
import numpy as np

X_train = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/penguin_X_train.csv")
X_test = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/penguin_X_test.csv")
y_train = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/penguin_y_train.csv")

train = pd.concat([X_train, y_train], axis=1) # y_train을 포함해서결측치 제거를 위해 합쳐줌

train.dropna(inplace=True)
train.reset_index(drop=True, inplace=True) # drop=True 옵션은 기존의 인덱스를 버리고 재배열해준다.

X_train = train.iloc[:,:-1]
y_train = train.iloc[:,-1]

X_train.describe()

# print(X_train.info())
# print(X_train)
# 수치형 변수 3개는 스케일링이 필요, 범주형 변수 3개에 대해서는 인코딩이 필요

X = pd.concat([X_train, X_test])

categoric_ = X.columns[:3] # 범주형 변수 컬럼명
numeric_ = X.columns[3:] # 수치형 변수 컬럼명

from sklearn.preprocessing import OneHotEncoder, MinMaxScaler

ohe = OneHotEncoder(handle_unknown = 'ignore')

# 범주형 변수는 One Hot Encoding
X_train_res = ohe.fit_transform(X_train[categoric_])
X_test_res = ohe.fit_transform(X_test[categoric_])

X_train_ohe = pd.DataFrame(X_train_res.todense(), columns = ohe.get_feature_names())
X_test_ohe = pd.DataFrame(X_test_res.todense(), columns = ohe.get_feature_names())

print(X_train_ohe)

# 범주형 변수를 One Hot Encoding으로 수치형으로 변환해줬으니 기존의 수치형 변수와 합친다.
X_train_fin = pd.concat([X_train[numeric_], X_train_ohe], axis=1)
X_test_fin = pd.concat([X_test[numeric_], X_test_ohe], axis=1)

from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train_fin, y_train, test_size=0.3, random_state=42)

scaler = MinMaxScaler()
scaler.fit(X_tr[numeric_])
# 수치형 변수는 Min Max Scaling
X_tr[numeric_] = scaler.transform(X_tr[numeric_])
X_val[numeric_] = scaler.transform(X_val[numeric_])
X_test_fin[numeric_] = scaler.transform(X_test_fin[numeric_])

print(X_train_fin)

from sklearn.linear_model import LinearRegression
import numpy as np
modelLR = LinearRegression()
modelLR.fit(X_tr, y_tr)

y_val_pred = modelLR.predict(X_val)
print(y_val_pred)

print(modelLR.intercept_) # 기울기 확인

coef = pd.Series(data=modelLR.coef_, index=X_train_fin.columns) # 계수 확인
print(coef.sort_values())

from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_val, y_val_pred)
rmse = mean_squared_error(y_val, y_val_pred, squared=False)
rmse_2 = np.sqrt(mse)
print('MSE : {0:.3f} , RMSE : {1:.3f}, RMSE : {2:.3f}'.format(mse, rmse, rmse_2)) # sqrt를 활용해도 rmse를 구할 수 있다.
```

```python
import pandas as pd
import numpy as np

X_train = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/penguin_X_train.csv")
X_test = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/penguin_X_test.csv")
y_train = pd.read_csv("/content/yemoonsaBigdata/datasets/Part2/penguin_y_train.csv")


print(X_train.info())
print(X_train)
```

```python
y_train # 몸무게를 예측하는 회귀분석
```

```python
train = pd.concat([X_train, y_train], axis=1) # y_train을 포함해서결측치 제거를 위해 합쳐줌
```

```python
train.dropna(inplace=True)
train.reset_index(drop=True, inplace=True) # drop=True 옵션은 기존의 인덱스를 버리고 재배열해준다.
```

```python
train
```

```python
# 다시 독립변수와 종속변수로 분리시킴

X_train = train.iloc[:,:-1]
y_train = train.iloc[:,-1]
```

```python
X_train.describe()
```

```python
# 수치형 변수 3개는 스케일링이 필요, 범주형 변수 3개에 대해서는 인코딩이 필요

X = pd.concat([X_train, X_test])

categoric_ = X.columns[:3] # 범주형 변수 컬럼명
numeric_ = X.columns[3:] # 수치형 변수 컬럼명

from sklearn.preprocessing import OneHotEncoder, MinMaxScaler

ohe = OneHotEncoder(handle_unknown = 'ignore')
scaler = MinMaxScaler()

# 범주형 변수는 One Hot Encoding
X_train_res = ohe.fit_transform(X_train[categoric_])
X_test_res = ohe.fit_transform(X_test[categoric_])

# 수치형 변수는 Min Max Scaling
X_train[numeric_] = scaler.fit_transform(X_train[numeric_])
X_test[numeric_] = scaler.fit_transform(X_test[numeric_])
```

```python
print(X_train_res), print(X_train[numeric_])
```

```python
X_train_ohe = pd.DataFrame(X_train_res.todense(), columns = ohe.get_feature_names())
X_test_ohe = pd.DataFrame(X_test_res.todense(), columns = ohe.get_feature_names())

print(X_train_ohe)

# 범주형 변수를 One Hot Encoding으로 수치형으로 변환해줬으니 기존의 수치형 변수와 합친다.
X_train_fin = pd.concat([X_train[numeric_], X_train_ohe], axis=1)
X_test_fin = pd.concat([X_test[numeric_], X_test_ohe], axis=1)

print(X_train_fin)
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train_fin, y_train, test_size=0.3)
```

```python
X_val
```

```python
from sklearn.linear_model import LinearRegression

modelLR = LinearRegression()
modelLR.fit(X_tr, y_tr)

y_val_pred = modelLR.predict(X_val)
print(y_val_pred)
```

```python
print(modelLR.intercept_) # 기울기 확인

coef = pd.Series(data=modelLR.coef_, index=X_train_fin.columns) # 계수 확인
print(coef.sort_values())
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_val, y_val_pred)
rmse = mean_squared_error(y_val, y_val_pred, squared=False)
rmse_2 = np.sqrt(mse)
print('MSE : {0:.3f} , RMSE : {1:.3f}, RMSE : {2:.3f}'.format(mse, rmse, rmse_2)) # sqrt를 활용해도 rmse를 구할 수 있다.
```
